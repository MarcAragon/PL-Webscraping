```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
def Limpieza_Goles(cadena):

  x=cadena.replace("(", "").replace(")", "").replace("clubes", "")
  
  for n in range(0,1001):
            if f"{n}" in x:
                x = x.replace(f"{n}", "")

  
  
  return x
```

```python
url = "https://es.wikipedia.org/wiki/Premier_League"
```

```python
page = requests.get(url)
```

```python
Page_Txt = BeautifulSoup(page.text, 'html')
```

```python
xTables = Page_Txt.find_all('table')
```

```python
Season_23_24_Table = xTables[5] #Encontrado a mano analizando uno por uno de los elementos de x
S23_24_Hmlt_Titles = Season_23_24_Table.find_all('th')


Table_Titles_S23_24 = []
for word in S23_24_Hmlt_Titles:
    Table_Titles_S23_24.append(word.text.strip())
Table_Titles_S23_24.pop(-1)
```

```python
Season_23_24 = pd.DataFrame(columns=Table_Titles_S23_24)
Season_23_24
```

```python
Table_Data_S_23_23 = Season_23_24_Table.find_all('tr')

for rows in Table_Data_S_23_23[1:-1]: #Como el elemento 0 es la informacion de las columnas, no se toma para el ciclo al añadirlo al datrame porque pondria error
    
    row = rows.find_all('td')  #Y arriba se excluye el -1 porque tampoco es informacion que entra al dataframe
    Row_Data = [] 
    
    for word in row: 
        
        x = word.text.strip()
        
        for n in range(0,101):
            if str(n) in x:
                x = x.replace(str(f"[{n}]"), "") #Los entrenadores tienen referencias asi que se quitan
        
        Row_Data.append(x)
    
    longitud = len(Season_23_24)
    Season_23_24.loc[longitud] = Row_Data
```

```python
for index, row in Season_23_24.iterrows(): #Ya que no se puede convertir directamente a int ya que hay un espacio en medio, se remueve
    row["Aforo"] = "".join(row["Aforo"].split()) 
```

```python
Goalscorers_Table = xTables[11]
Goalscorers_Html_Titles = Goalscorers_Table.find_all('th')
```

```python
Goalscorers_Titles = []
for word in Goalscorers_Html_Titles:
    Goalscorers_Titles.append(word.text.strip())
```

```python
Goalcorers_Df = pd.DataFrame(columns=Goalscorers_Titles)
Goalcorers_Df
```

```python
Goalcorers_Df = Goalcorers_Df.rename(columns={"G.": "Goles",
                                              "Part.": "Partidos",
                                              "Prom.": "Promedio",
                                              "(Equipo debut)": "Equipo_Debut",
                                              "Otros clubes": "Otros_Clubes"})
```

```python
Goalscorers_Table_Data = Goalscorers_Table.find_all("tr")
```

```python
for rows in Goalscorers_Table_Data[1:]: #Como el elemento 0 es la informacion de las columnas, no se toma para el ciclo al añadirlo al datrame porque pondria error
    
    row = rows.find_all('td')  
    Row_Data = [] 
    
    for word in row: 
        
        x = word.text.strip()
        
        for n in range(0,101): #Eliminar referencias
            if str(n) in x:
                x = x.replace(str(f"[{n}]"), "") 
        
        Row_Data.append(x)
    
    longitud = len(Goalcorers_Df)
    Goalcorers_Df.loc[longitud] = Row_Data
```

```python
Goalcorers_Df
```

```python
Goalcorers_Df.set_index("Pos.", inplace=True)
Goalcorers_Df
```

```python
Goalcorers_Df["Goles_Equipo_Debut"] = pd.NA
```

```python
Goalcorers_Df
```

```python
Goles_Debut = []

for index, row in Goalcorers_Df.iterrows(): #Se saca la informacion de los goles en equipo debut
    number = ""
    
    for i in row["Equipo_Debut"]:
        if i.isdigit():
            number += i #Se dejan todos los numeros, se dejan meten en una cadena string y a una lista
    
    Goles_Debut.append(number)
```

```python
Goalcorers_Df["Goles_Equipo_Debut"] = Goles_Debut
```

```python
Goalcorers_Df
```

```python
Goalcorers_Df["Equipo_Debut"] = Goalcorers_Df["Equipo_Debut"].apply(Limpieza_Goles)
```

```python
Goalcorers_Df
```

```python
Goles_Otros_Clubs = []

for index, row in Goalcorers_Df.iterrows(): #Se saca la informacion de los goles en otros equipos
    
    number = 0
    for n in range(0,1001):
        if f"({n})" in row["Otros_Clubes"]: #Se suman los numeros entre parentesis para cada fila y se añaden a una lista
            number += n
    
    Goles_Otros_Clubs.append(number)
```

```python
Goalcorers_Df["Goles_Otros_Clubes"] = Goles_Otros_Clubs
```

```python
Goalcorers_Df
```

```python
Goalcorers_Df["Otros_Clubes"] = Goalcorers_Df["Otros_Clubes"].apply(Limpieza_Goles)
```

```python
Goalcorers_Df.head()
```

```python
Goalcorers_Df = Goalcorers_Df.reindex(columns=['Jugador', 'Goles', 'Partidos', 'Promedio', 'Debut', 'Equipo_Debut', 'Goles_Equipo_Debut', 'Otros_Clubes', 'Goles_Otros_Clubes'])
```

```python
Goalcorers_Df
```

```python
Goalcorers_Df['Goles'] = Goalcorers_Df['Goles'].astype(int)
Goalcorers_Df['Partidos'] = Goalcorers_Df['Partidos'].astype(int)
Goalcorers_Df['Promedio'] = Goalcorers_Df['Promedio'].astype(float)
Goalcorers_Df['Goles_Equipo_Debut'] = Goalcorers_Df['Goles_Equipo_Debut'].astype(int)
Goalcorers_Df['Goles_Otros_Clubes'] = Goalcorers_Df['Goles_Otros_Clubes'].astype(int)
```

```python
Debut_Goalscorers = Goalcorers_Df.groupby(by="Equipo_Debut")
```

```python
print("Goles anotados por jugadores que debutaron en estos equipos")
Debut_Goalscorers['Goles'].sum()
```

```python
print("Goles anotados en los equipos debut de los jugadores")
Debut_Goalscorers['Goles_Equipo_Debut'].sum()
```
